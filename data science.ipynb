```python
import numpy as np
import pandas as pd

np.random.seed(23)

data = {
    "x" : np.arange(1,10),
    "y" : np.random.randint(15,100,size=9)
}

df = pd.DataFrame(data)
print(df)

import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))
plt.plot(df)

plt.xlabel("days")
plt.ylabel("sales")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))
plt.plot(df["x"], df["y"])

plt.xlabel("days")
plt.ylabel("sales")

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(df["x"], df["y"], marker="o", color='red')

plt.xlabel("X Values")
plt.ylabel("Random Y Values")
plt.title("plot")
plt.grid(True)
plt.show()

plt.show()
```

```python
import numpy as np
x = np.arange(1,11)
y = np.random.randint(10, 100, size=10)
data= np.random.randn(100)
print(x)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.plot(x,y, marker='x', linestyle='--',color='red')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line plot")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.bar(x,y,color='grey')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line plot")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.scatter(x,y,color='red')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter plot")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.barh(x,y,color='red')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Horizontal Bar plot")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.hist(x,y,color='red')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Histogram plot")
plt.grid(False)
plt.show()

```
