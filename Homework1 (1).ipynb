```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 11, 1)
y = x**2
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='--', color='red')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Graph of y = x^2")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
products = ['A', 'B', 'C', 'D']
sales = [40, 70, 30, 90]
plt.figure(figsize=(6, 4))
plt.bar(products, sales, color='green')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales of Products A, B, C, and D")
plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = np.arange(1, 11)
y = x ** 2
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y, marker='o',color='purple')
plt.title("Line Plot")
plt.grid(True)
plt.subplot(2, 2, 2)
plt.bar(['A','B','C','D'], [40, 70, 30, 90],color='purple')
plt.title("Bar Plot")
plt.subplot(2, 2, 3)
plt.scatter(x, y,marker='o',color='purple')
plt.title("Scatter Plot")
plt.grid(True)
plt.subplot(2, 2, 4)
plt.hist(np.random.randn(100),color='purple')
plt.title("Histogram")

plt.show()
```

```python

data = np.random.randn(200)
plt.figure(figsize=(6, 4))
plt.hist(data, bins=20, color='orange')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
